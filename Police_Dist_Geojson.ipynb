# Get GeoJSON data for the Police Districts

```python
# Import our dependencies
import requests
import time
from datetime import datetime
import pandas as pd
import numpy as np
import geopandas as gpd
import json
```

```python
gdf = gpd.read_file('https://data.cityofchicago.org/api/geospatial/fthy-xz3r?method=export&format=Shapefile')
gdf
```

```python
# Rename columns
gdf.rename(columns = {'dist_num':'Police_Dist'}, inplace = True)
gdf.head(1)
```
